```python
#preprocessing
data.drop(['Loan_ID'],axis=1,inplace=True)
```

```python
data.info()
```

```python
# Handling Null values
for i in ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']:
  data[i].fillna(data[i].mean(), inplace=True)

for i in ['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']:
  data[i].fillna(data[i].mode()[0], inplace=True)
```

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.drop('Loan_Status',axis=1), data['Loan_Status'], test_size=.3,random_state =123, stratify = data['Loan_Status'])
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```
