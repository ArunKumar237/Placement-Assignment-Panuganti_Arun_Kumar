```python
df['price']=df['price'].fillna(df['price'].mean()) # filling null values with mean
```

```python
df['datetime'] = df['datetime'].astype('datetime64[ns]')
```

```python
df = df[df['datetime'].dt.month == 11]
```

```python
df = df.sample(n=50000)
```

```python
df.isna().sum()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
import folium

df = df[['latitude', 'longitude', 'price']]

# Perform unsupervised learning to predict high booking areas
X = df[['latitude', 'longitude']]  # Select relevant features for clustering
kmeans = KMeans(n_clusters=3)  # Specify the number of clusters
kmeans.fit(X)
df['cluster'] = kmeans.labels_  # Assign cluster labels to each data point

# Perform supervised learning to predict prices for locations
X_train = df[['latitude', 'longitude']]  # Input features for training
y_train = df['price']  # Target variable for training
reg_model = LinearRegression()  # Choose a regression model
reg_model.fit(X_train, y_train)  # Train the model

```

```python
# Display the data on a map
center_lat = np.mean(df['latitude'])  # Calculate center latitude
center_lon = np.mean(df['longitude'])  # Calculate center longitude
map_obj = folium.Map(location=[center_lat, center_lon], zoom_start=12)  # Create a Folium map

# Add clustered locations to the map
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=5,
        color='blue' if row['cluster'] == 0 else 'green' if row['cluster'] == 1 else 'red',  # Assign colors based on cluster label
        fill=True,
        fill_color='blue' if row['cluster'] == 0 else 'green' if row['cluster'] == 1 else 'red',
        fill_opacity=0.6
    ).add_to(map_obj)

# Add predicted price information to the map
for _, row in df.iterrows():
    predicted_price = reg_model.predict([[row['latitude'], row['longitude']]])
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        icon=folium.Icon(color='gray', icon='info-sign'),
        popup=f"Price: {predicted_price[0]:.2f}"
    ).add_to(map_obj)

# Display the map
map_obj.save('map.html')  # Save the map as an HTML file
map_obj  # Display the map in Jupyter Notebook or IDE

```
