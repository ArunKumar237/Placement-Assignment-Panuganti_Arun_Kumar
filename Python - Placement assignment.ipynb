```python
class Base:
    def __init__(self):
        self.dictionary = {}
    
    def count_ele(self, elements):
        if elements[-1] == '.':
            elements = elements[0:len(elements) - 1]

        if elements in self.dictionary:
            self.dictionary[elements] += 1
        else:
            self.dictionary.update({elements: 1})
            
    def freq(self, dic):
        wordlist = []
        if dic[0][1] > 1:
            for i in dic:
                if i[1] == dic[0][1]:
                    wordlist.append(i[0])
            print('\nresult:\n\t',sorted(wordlist, reverse=True)[0],':',len(sorted(wordlist, reverse=True)[0]))
        else:
            print('\nresult:\n\t','No repetative words')

    def get_highest_freq(self):
        sentence = input()
        lst = sentence.split()
        
        for i in lst:
            self.count_ele(i)
        sorted_dic = sorted(self.dictionary.items(), key=lambda x:x[1], reverse=True)
        self.freq(sorted_dic)
    
```

```python
import pandas as pd
import requests
import re

response = requests.get('http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes')
json_data = response.json()
```

```python
json_data.keys()
```

```python
dataframe = pd.DataFrame(json_data['_embedded'])
dataframe
```

```python
dataframe = dataframe['episodes'].apply(pd.Series) # expanding the 'episodes' column
```

```python
def remove_tags(string):
    result = re.sub('<.*?>','',string)
    return result

a = dataframe[['id','url','name','season','number','type','airdate','airtime','runtime']]
b = dataframe['rating'].apply(pd.Series) # expanding the 'rating' column
b = b.rename(columns={'average': 'average_rating'})
c = dataframe['summary'].apply(remove_tags) # removing html tags of summary column
d = dataframe['image'].apply(pd.Series) # expanding column
df = pd.concat([a,b,c,d], axis=1) # concatinating the formated dataframes
df = df.set_index('id') # setting id as an index
df.to_csv('tvshow.csv')
df
```

```python
import nltk
import spacy
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
nlp = spacy.load('en_core_web_sm')

def filter():
    dic = {}
    dic2 = {}
    noun = 0
    prn = 0
    adj = 0
    vb = 0
    py_txt = input('your text here: \n\t')
#     py_txt = 'your text here: In the example below, we have seen that we have imported the nltk module. After importing the nltk module, we imported the stopwords module using the nltk.corpus. Then we have to import the word_tokenize and sent_tokenize module using nltk.tokenize module. After importing all the modules, we are calling the set method. After calling the set method, we have defined the sentence to use in pos tagging. After defining the sentence, we are calling the sent_tokenize and word_tokenize methods. Then we have to call the nltk.pos_tag method to use the nltk pos tag in our code. At the time of calling this method, we have created the object. After creating the object, we have printed this in output.'
    py_sword = set(stopwords.words ('english'))
    py_token = sent_tokenize (py_txt)
    
    for i in py_token:
        # POS tagging using spaCy
        doc = nlp(i)
        spacy_pos_tags = [(token.text, token.pos_) for token in doc]
    
    for i,j in spacy_pos_tags:
        if j in ['NOUN','PRON','VERB','ADJ']:
            dic2[i]=j
            if j=='NOUN':
                noun += 1
                dic['Noun']=noun
            if j=='PRON':
                prn += 1
                dic['Pronoun']=prn
            if j=='VERB':
                vb += 1
                dic['Verb']=vb
            if j=='ADJ':
                adj += 1
                dic['Adjective']=adj
    print('')
    print(dic2)
    print('')
    print(dic)
            
```
